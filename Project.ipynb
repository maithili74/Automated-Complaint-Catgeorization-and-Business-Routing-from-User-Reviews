```python
# Text preprocessing
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
import re
def cleaned_text(text):
  text=text.lower()
  text=re.sub(r'<.*?>', '', text)             # Remove HTML
  text = re.sub(r'\s+', ' ', text)              # Normalize whitespace
  text = re.sub(r'https?://\S+|www\.\S+', '', text)  # Remove URLs
  words = text.split()
  words = [word for word in words if word not in stop_words]  # Remove stopwords
  return ' '.join(words)

df['cleaned_review']=df['content'].apply(cleaned_text)
df.head()
```

```python
df=df.drop(columns=['content'])
df.head()
```

```python
# Keep only negative and neutral reviews
df = df[df['score'].isin([1, 2, 3])]

# Create a new column for sentiment label
df['sentiment'] = df['score'].apply(lambda x: 'negative' if x in [1, 2] else 'neutral')
```

```python
from bertopic import BERTopic

docs = df['cleaned_review'].tolist()


topic_model = BERTopic(language="english", verbose=True)


topics, probs = topic_model.fit_transform(docs)
```

```python

df['topic'] = topics
df['topic_prob'] = probs
```

```python
df.head()
```

```python
df['topic'].value_counts()
```

```python
topic_info = topic_model.get_topic_info()
topic_info.head(10)
```

```python
topic_names = {
    0: "Widget Functionality Issues",
    1: "Pro Version / Tree Coins Complaints",
    2: "Habit Tracking Inaccuracies",
    3: "Calendar & Holiday Sync Problems",
    4: "Reminder & Alarm Notification Failures",
    5: "Wunderlist Integration / Microsoft Sync Issues",
    6: "Notification Setting Bugs",
    7: "Task Repetition & Recurrence Bugs",
    8: "Subscription & Paywall Complaints",
}


df['topic_name'] = df['topic'].apply(lambda x: topic_names.get(x, "Miscellaneous / Other Issues"))
```

```python
df.head()
```

```python
df.head(10)
```

```python
df_cleaned = df[['score', 'cleaned_review', 'sentiment', 'topic_name', 'topic','topic_prob']]
```

```python
df_cleaned.head()
```

```python
TOP_N = 5


topic_counts = df_cleaned['topic_name'].value_counts().reset_index()
topic_counts.columns = ['topic_name', 'review_count']


topic_counts['percent_of_total'] = (topic_counts['review_count'] / len(df)) * 100


top_n_issues = topic_counts.head(TOP_N)


top_n_total = top_n_issues['review_count'].sum()
top_n_coverage_percent = (top_n_total / len(df)) * 100


print("Top N Issues:")
print(top_n_issues)

print(f"\nTotal reviews covered by top {TOP_N} issues: {top_n_total}")
print(f"Percentage of total reviews covered by top {TOP_N} issues: {top_n_coverage_percent:.2f}%")
```

```python
sentiment_dist = df_cleaned.groupby(['topic_name', 'sentiment']).size().reset_index(name='count')


sentiment_pivot = sentiment_dist.pivot(index='topic_name', columns='sentiment', values='count').fillna(0)


sentiment_pivot['total_reviews'] = sentiment_pivot.sum(axis=1)
sentiment_pivot['negative_ratio'] = sentiment_pivot.get('negative', 0) / sentiment_pivot['total_reviews']
sentiment_pivot = sentiment_pivot.sort_values(by='negative_ratio', ascending=False)

print(sentiment_pivot)
```

```python
routing_map = {
    "Widget Functionality Issues": "Tech",
    "Pro Version / Tree Coins Complaints": "Billing",
    "Habit Tracking Inaccuracies": "Product",
    "Calendar & Holiday Sync Problems": "Tech",
    "Reminder & Alarm Notification Failures": "Tech",
    "Wunderlist Integration / Microsoft Sync Issues": "Tech",
    "Notification Setting Bugs": "Tech",
    "Task Repetition & Recurrence Bugs": "Product",
    "Subscription & Paywall Complaints": "Billing",
    "Miscellaneous / Other Issues": "Support"
}


df_cleaned['assigned_team'] = df_cleaned['topic_name'].map(routing_map).fillna('Support')


team_distribution = df_cleaned['assigned_team'].value_counts().reset_index()
team_distribution.columns = ['team', 'review_count']

team_distribution['percent_of_total'] = (team_distribution['review_count'] / len(df_cleaned)) * 100


print(team_distribution)
```
